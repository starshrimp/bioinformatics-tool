## Merge EM + Clinical Metadata
Merging the datasets on the "title" column from the clinical metadataset.

```python
expr = em.T  

meta = gse.phenotype_data.copy()
meta.index = meta['title'].str.strip()  

common_ids = expr.index.intersection(meta.index)
print(f"{len(common_ids)} matched samples")

expr_common = expr.loc[common_ids]
meta_common = meta.loc[common_ids]

combined = expr_common.merge(meta_common, left_index=True, right_index=True)
print(combined.shape)
```

## Overview and Sanity Checks

### Constant Columns

I noticed there were some columns that only contained one unique value in the whole dataset, so I delete these columns before saving the dataset to csv to save computational time and power.

```python
# Identify columns w only 1 unique value
constant_cols = [col for col in combined.columns if combined[col].nunique(dropna=False) == 1]

# printing them
print(f"{len(constant_cols)} constant columns:")
for col in constant_cols:
    print(f" - {col}")

```

There were a lot of constant columns (2262 constant columns). This is likely due to genes being measured that may not be expressed in breast tissue at all or might be below detection limit. To test this, I want to have a closer look: 

```python
# Check constant value of first few constant columns
for col in constant_cols[:5]:
    print(f"{col}: {combined[col].unique()}")
```

This value is the log2(0.1), which could be a placeholder for no expression.

```python
numeric_constants = combined[constant_cols].select_dtypes(include='number').columns
print(f"{len(numeric_constants)} constant gene expression columns")
```

2236 out of 2262 constant columns were gene columns. Now the next step is to analyze, whether they all have zero expression.

```python
import numpy as np

log_zero = np.log2(0.1)

numeric_constants = combined[constant_cols].select_dtypes(include='number').columns


all_match_log_zero = []
for col in numeric_constants:
    unique_val = combined[col].iloc[0]
    if np.isclose(unique_val, log_zero):
        all_match_log_zero.append(col)

# Print result
print(f"{len(numeric_constants)} constant numeric columns (likely gene expression)")
print(f"{len(all_match_log_zero)} of them have value == log2(0.1) ≈ {log_zero:.5f}")
```

## Preprocessing
Problems identified:
highly right-skewed distribution, dominant left peak at -3.3 (= zero expression), with a long right tail
- columns with only 1 or a few samples where gene is expressed
- constant columns - contribute nothing due to same expression (zero expression) for all samples -> remove
- median-centering was not applied
- right-skewness

### Dropping Constants & Median Centering

```python
# Step 1: Drop constant columns (already done)
cleaned_combined = combined.drop(columns=constant_cols)

# Step 2: Split numeric and non-numeric
non_numeric = cleaned_combined.select_dtypes(exclude='number')
cleaned_combined_numeric = cleaned_combined.select_dtypes(include='number')

# Step 3: Median centering per gene (per column!)
median_centered = cleaned_combined_numeric.sub(cleaned_combined_numeric.median(axis=0), axis=1)


```

```python
import matplotlib.pyplot as plt

# Flatten the matrix to 1D for plotting
values = median_centered.values.flatten()

# Plot histogram
plt.figure(figsize=(8, 5))
plt.hist(values, bins=100, color='skyblue', edgecolor='black')
plt.axvline(x=0, color='red', linestyle='--', label='Zero')
plt.title("Distribution of Median-Centered Expression Values")
plt.xlabel("Expression Value")
plt.ylabel("Frequency")
plt.grid(True, linestyle='--', alpha=0.6)
plt.legend()
plt.tight_layout()
plt.show()
```

### Z-Scoring

```python
zscored = median_centered.sub(median_centered.mean(axis=0), axis=1).div(median_centered.std(axis=0), axis=1)
```

```python
import matplotlib.pyplot as plt

# Flatten all z-scored values into a 1D array
all_values = zscored.values.flatten()

# Plot histogram
plt.figure(figsize=(8, 5))
plt.hist(all_values, bins=100, color='steelblue', edgecolor='black')
plt.xlabel("Z-scored expression value")
plt.ylabel("Frequency")
plt.title("Distribution of Z-scored Gene Expression Values")
plt.axvline(x=0, color='red', linestyle='--', label='Zero')
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```

#### Exploring Outliers

```python
import numpy as np

# Choose a threshold for outliers
threshold = 5

# Find boolean matrix where values are outliers
outlier_mask = (zscored.abs() > threshold)

# Count number of outliers per gene (column-wise since samples are rows)
outlier_counts = outlier_mask.sum(axis=0).sort_values(ascending=False)

# Top genes with the most outlier values
outlier_counts.head(10)
```

```python
from scipy.stats import chi2_contingency
import pandas as pd

# Define thresholds
high_thresh = 5
low_thresh = -5

# Add flags
recombined["ERAL1_high"] = recombined["ERAL1"] > high_thresh
recombined["PRELID2_low"] = recombined["PRELID2"] < low_thresh

# Identify categorical columns
meta_cols = recombined.select_dtypes(exclude='number').columns

# Collect results
results = []

for col in meta_cols:
    try:
        df = recombined[[col, "ERAL1_high", "PRELID2_low"]].dropna()
        if df[col].nunique() > 1:
            # ERAL1
            ct1 = pd.crosstab(df["ERAL1_high"], df[col])
            if ct1.shape[0] > 1 and ct1.shape[1] > 1:
                _, p1, _, _ = chi2_contingency(ct1)
                results.append(("ERAL1_high", col, p1))
            # PRELID2
            ct2 = pd.crosstab(df["PRELID2_low"], df[col])
            if ct2.shape[0] > 1 and ct2.shape[1] > 1:
                _, p2, _, _ = chi2_contingency(ct2)
                results.append(("PRELID2_low", col, p2))
    except Exception as e:
        continue

# Filter and display
sig = pd.DataFrame(results, columns=["GeneFlag", "MetadataColumn", "P-value"])
sig = sig[sig["P-value"] < 0.05].sort_values("P-value")
sig
```
