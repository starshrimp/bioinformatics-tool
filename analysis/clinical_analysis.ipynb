```python
import pandas as pd

# Load the dataset
file_path = '/Users/sarah/Code/bioinformatics-tool/backend/data/GSE96058_clinical_metadata_cleaned.csv'
df = pd.read_csv(file_path)

# Display basic information
print("Shape of the dataset:", df.shape)
print("\nColumns:", df.columns.tolist())
print("\nFirst 5 rows:")
print(df.head())

# Summary statistics
print("\nSummary statistics:")
print(df.describe(include='all'))

# Check for missing values
print("\nMissing values per column:")
print(df.isnull().sum())
```

```python
import matplotlib.pyplot as plt

# Calculate the counts for each PAM50 subtype
pam50_counts = {
  'Basal': df['pam50 subtype__Basal'].sum(),
  'Her2': df['pam50 subtype__Her2'].sum(),
  'LumA': df['pam50 subtype__LumA'].sum(),
  'LumB': df['pam50 subtype__LumB'].sum(),
  'Normal': df['pam50 subtype__Normal'].sum()
}

# Create the bar plot
plt.figure(figsize=(8, 5))
plt.bar(pam50_counts.keys(), pam50_counts.values(), color='skyblue')
plt.xlabel('PAM50 Subtype')
plt.ylabel('Count')
plt.title('Distribution of PAM50 Subtypes')
plt.show()
```
